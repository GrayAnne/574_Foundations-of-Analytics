```python
import pandas as pd
import matplotlib.pyplot as plt

# Read film.csv into a DataFrame
film_df = pd.read_csv('film.csv', sep = '|', quotechar="'")

# Read categories.xlsx into a DataFrame
cate_df = pd.read_excel('categories.xlsx')
```

```python
film_df
```

```python
cate_df
```

```python
merge = pd.merge(film_df, cate_df, on='category_id')
merge
```

### section4

```python
counts = merge["category_name"].value_counts()
fig = plt.figure(figsize = (20, 10))
plt.bar(counts.index, counts.iat, color ='cornflowerblue',width = 0.5) 
plt.xlabel("Film Category")
plt.ylabel("Number of Films")
plt.title("Number of Films by Category")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
duration = merge.groupby('category_name')['rental_duration']
f=plt.boxplot([group for name, group in duration], labels = duration.groups.keys(), patch_artist=True,)
for i in f['boxes']:i.set_facecolor("cornflowerblue")
plt.xlabel("Film Category")
plt.ylabel('Rental Duration')
plt.title('Rental Duration by Category')
plt.show()
```
