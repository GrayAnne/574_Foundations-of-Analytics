```python
'''1. Suppose tomorrow will be mild, rainy, and windy, with high humidity. Should we play golf 
tomorrow? Show how the answer was calculated.

P(Golf | rainy, mild, High, True) ∝P( rainy | Golf) * P( mild | Golf) * P( High | Golf) * P( True | Golf) * P(Golf)
P(!Golf | rainy, mild, High, True) ∝P( rainy | !Golf) * P( mild | !Golf) * P( High | !Golf) * P( True | !Golf) * P(!Golf)
'''

Golf_Tomorrow = 2/9 * 4/9 * 3/9 * 3/9 * 9/14
NoGolf_Tomorrow = 3/5 * 2/5 * 4/5 *3/5 * 5/14


if Golf_Tomorrow > NoGolf_Tomorrow:
    decision = "Play Golf"
else:
    decision = "Don't Play Golf"

print(Golf_Tomorrow)
print(NoGolf_Tomorrow)
print(f"The decision is to '{decision}'")
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('PlayGolf.csv')
df
```

```python
'''
1. Temperature and Outlook are ordinal variables. Instead of converting them to dummy variables, 
they need to be recoded as ordinal variables. Use this link for guidance
'''
Temperature_mapper = {"Cool":1, "Mild":2, "Hot":3}
Outlook_mapper = {"Sunny":1, "Overcast":2, "Rainy":3}
df["Temperature"] = df["Temperature"].replace(Temperature_mapper)
df["Outlook"] = df["Outlook"].replace(Outlook_mapper)
print(df)
```

```python
'''
2. The other variables need to be recoded to binary variables. Because Windy is Boolean, it does 
not need to be recoded
'''
df = pd.get_dummies(df, columns = ['Humidity']) 
  
display(df) 
```

```python
'''
3. Fit the data using CategoricalNB.
'''
from sklearn.naive_bayes import CategoricalNB

X = df.drop('PlayGolf', axis=1)
y = df['PlayGolf']
clf = CategoricalNB()
clf.fit(X, y)
clf.predict(X)
```

```python
'''
4. Using the data set, PlayGolfNext.csv, use your Naïve Bayes model to predict the next few days.
Which days should you play golf?
'''
df2 = pd.read_csv('PlayGolfNext.csv')

Temperature_mapper = {"Cool":1, "Mild":2, "Hot":3}
Outlook_mapper = {"Sunny":1, "Overcast":2, "Rainy":3}
df2["Temperature"] = df2["Temperature"].replace(Temperature_mapper)
df2["Outlook"] = df2["Outlook"].replace(Outlook_mapper)
df2 = pd.get_dummies(df2, columns = ['Humidity']) 

test = df2.drop(['Day'], axis=1)
df2['PlayGolf'] = clf.predict(test)

df2
```
